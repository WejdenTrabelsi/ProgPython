```python
import sqlite3
from tkinter import *
```

```python
# Fenêtre principale
fen = Tk()
fen.title("Management Students")
fen.geometry("385x244")
```

```python

# Labels
lbl1 = Label(fen, text="Management Students")
lbl1.place(x=90, y=20)

lbl2 = Label(fen, text="Name Student :")
lbl2.place(x=30, y=70)

lbl3 = Label(fen, text="E_Mail :")
lbl3.place(x=30, y=105)

lbl4 = Label(fen, text="Age :")
lbl4.place(x=30, y=130)

# Entrées
ent1 = Entry(fen)
ent1.place(x=150, y=70)

ent2 = Entry(fen)
ent2.place(x=150, y=102)

ent3 = Entry(fen)
ent3.place(x=150, y=130)

# Zone résultat
result = Text(fen, width=50, height=11)
result.place(x=30, y=220)
```

```python

# Fonction connexion DB
def connexiondb():
    try:
        conn = sqlite3.connect("DBstudents.db")
        cur = conn.cursor()
        return conn, cur
    except:
        print("Erreur de connexion")

# Create Table
def create_table():
    conn, cur = connexiondb()
    cur.execute("""
        CREATE TABLE IF NOT EXISTS Students(
            idstd INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            email TEXT NOT NULL UNIQUE,
            age INTEGER NOT NULL
        )
    """)
    conn.commit()
    conn.close()

# Insert
def insert_data():
    name = ent1.get()
    email = ent2.get()
    age = ent3.get()

    conn, cur = connexiondb()
    cur.execute(
        "INSERT INTO Students(name, email, age) VALUES (?, ?, ?)",
        (name, email, age)
    )
    conn.commit()
    conn.close()

# Display Data
def display_data():
    result.delete(1.0, END)
    conn, cur = connexiondb()
    cur.execute("SELECT name, email, age FROM Students")
    rows = cur.fetchall()

    for row in rows:
        result.insert(END, "Nom : " + str(row[0]) + "\n")
        result.insert(END, "Email : " + str(row[1]) + "\n")
        result.insert(END, "Age : " + str(row[2]) + "\n")
        result.insert(END, "************\n")

    conn.close()

# Clear
def clear_fields():
    ent1.delete(0, END)
    ent2.delete(0, END)
    ent3.delete(0, END)
    result.delete(1.0, END)

# Boutons
btn1 = Button(fen, text="Insert", command=insert_data)
btn1.place(x=50, y=182)

btn2 = Button(fen, text="Close", command=fen.destroy)
btn2.place(x=155, y=182)

btn3 = Button(fen, text="Clear", command=clear_fields)
btn3.place(x=260, y=182)

btn4 = Button(fen, text="Create Table", command=create_table)
btn4.place(x=320, y=68)

btn5 = Button(fen, text="Display Data", command=display_data)
btn5.place(x=30, y=410)

fen.mainloop()
```
