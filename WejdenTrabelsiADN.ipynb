```python
import tkinter as tk
from tkinter import PhotoImage
from random import choice
```

```python

def radom_ADN(size):
    adn = ["A", "T", "C", "G"]
    ADN = ''
    for i in range(size):
        ADN += choice(adn)
    return ADN

def proportion(brin):
    D1 = {}
    D2 = {}
    for elt in set(brin):
        count = brin.count(elt)
        D1[elt] = count
        prc = round((count / len(brin) * 100), 2)
        D2[elt] = prc
    return D1, D2

def transform_arm(brin):
    if 'T' in brin:
        return "ARN : " + brin.replace("T", "U")
    else:
        return "Pas de transformation pour : " + brin

def composit_seq(brin):
    L = list(brin)
    for i in range(len(L)):
        if L[i] == "T":
            L[i] = "A"
        elif L[i] == "A":
            L[i] = "T"
        elif L[i] == "C":
            L[i] = "G"
        elif L[i] == "G":
            L[i] = "C"
    return "".join(L)
```

```python

root = tk.Tk()
root.title("Les Structures Des Données")
root.geometry("430x500")
root.resizable(False, False)

bg_color = "#b3d9ff"
root.configure(bg=bg_color)
```

```python
title_label = tk.Label(
    root,
    text="Manipuler Un Brin D'ADN",
    font=("Arial", 16, "bold"),
    bg=bg_color
)
title_label.pack(pady=(10, 5))
```

```python
subtitle = tk.Label(
    root,
    text="Donner La longueur Du Brin D'ADN :",
    font=("Arial", 12, "bold"),
    bg=bg_color
)
subtitle.pack(pady=(5, 2))
```

```python

entry_size = tk.Entry(root, width=30, justify="center", font=("Arial", 12))
entry_size.pack(pady=5)
```

```python
btn_frame = tk.Frame(root, bg=bg_color)
btn_frame.pack(pady=10)

btn_style = {"font": ("Arial", 12), "width": 12, "bg": "#e6e6e6", "relief": "raised"}
```

```python
tk.Button(btn_frame, text="Run1", command=lambda: run_generate(), **btn_style).grid(row=0, column=0, padx=5, pady=5)
tk.Button(btn_frame, text="Run2", command=lambda: run_prop(), **btn_style).grid(row=0, column=1, padx=5, pady=5)
tk.Button(btn_frame, text="Run3", command=lambda: run_arn(), **btn_style).grid(row=1, column=0, padx=5, pady=5)
tk.Button(btn_frame, text="Run4", command=lambda: run_comp(), **btn_style).grid(row=1, column=1, padx=5, pady=5)
```

```python
output = tk.Text(root, height=12, width=50, font=("Arial", 11), bg="white")
output.pack(pady=10)

ADN_global = ""
```

```python

def run_generate():
    global ADN_global
    size = entry_size.get()
    output.delete("1.0", tk.END)
    if size.isdigit():
        ADN_global = radom_ADN(int(size))
        output.insert(tk.END, f"Brin ADN généré : {ADN_global}")
    else:
        output.insert(tk.END, "Veuillez entrer un nombre valide.")

def run_prop():
    if ADN_global:
        x, y = proportion(ADN_global)
        output.delete("1.0", tk.END)
        output.insert(tk.END, f"Occurrences : {x}\n\nPourcentages : {y}")
    else:
        output.delete("1.0", tk.END)
        output.insert(tk.END, "Veuillez d'abord générer un brin.")

def run_arn():
    if ADN_global:
        output.delete("1.0", tk.END)
        output.insert(tk.END, transform_arm(ADN_global))
    else:
        output.delete("1.0", tk.END)
        output.insert(tk.END, "Veuillez d'abord générer un brin.")

def run_comp():
    if ADN_global:
        output.delete("1.0", tk.END)
        output.insert(tk.END, "Brin complémentaire : " + composit_seq(ADN_global))
    else:
        output.delete("1.0", tk.END)
        output.insert(tk.END, "Veuillez d'abord générer un brin.")
```

```python
tk.Button(root, text="Exit", command=root.destroy, font=("Arial", 12),
          width=10, bg="#e6e6e6").pack(pady=5)

root.mainloop()
```
