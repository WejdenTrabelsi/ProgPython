```python
import requests
```

```python
URL = "https://fr.wikipedia.org/wiki/Liste_des_capitales_du_monde"
HEADERS ={
    'User-Agent':'MonScrapperBot/1.0 (+https://votre_site.com)'
    

}
```

```python
try:
    reponse = requests.get(URL, headers=HEADERS)
    reponse.raise_for_status()  
    contenu = reponse.text
    html_content = reponse.content
except Exception as e:
    print(f"Une erreur s'est produite lors de la requÃªte : {e}")
    exit()
```

```python
print('Code=', reponse.status_code)
print(reponse.content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
title=soup.find('title') 
print(title)
```

```python
print("Title of the page:",title.text)
```

```python
try:
    table = soup.find('table', {'class': 'wikitable sortable'})
    if not table:
        print("erreur : la table des capitales n'a pas Ã©tÃ© trouvÃ©e.")
        exit() #u can add the code here but we're deviding it into parts
except AttributeError as e:
    print(f"Erreur lors de l'analyse du HTML : {e}")
```

```python
print(table)
```

```python
capitales_data =[]
for row in table.find_all('tr')[1:]:
    cells=row.find_all('td')
    if len(cells) >=2:
        pays=cells[0].text.strip()
        capitale=cells[1].text.strip()
        capitales_data.append({
            'pays': pays,
            'capitale': capitale
        })
#print(pays)
#print(capitale)
print(capitales_data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(capitales_data)
```
