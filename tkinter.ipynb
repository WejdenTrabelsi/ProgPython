```python
%gui tk

from tkinter import *
from math import pi

Fen = Tk()

Fen.title("Calcul de surface de cercle")
Fen.geometry("400x200")
Fen.configure(bg="#f0f0f0")


rayon_lbl = Label(Fen, text="Rayon (r):", font=('ariel', 12), bg='#f0f0f0')
rayon_lbl.place(x=5, y=10)

rayon_entry = Entry(Fen, width=15, font=('ariel', 12))
rayon_entry.place(x=100, y=10)
rayon_entry.focus_set()


resultat_var = StringVar()
resultat_var.set("Surface : en attente...")

resultat_label = Label(Fen, textvariable=resultat_var, font=('ariel', 12), bg="#f0f0f0")
resultat_label.place(x=5, y=150)


class Cercle:
    def __init__(self, rayon):
        self.rayon = rayon

    def calculer_surface(self):
        if self.rayon < 0:
            raise ValueError("Le rayon ne peut pas être négatif")
        return pi * (self.rayon ** 2)



def calculer():
    try:
        rayon_str = rayon_entry.get().strip()
        if not rayon_str:
            resultat_var.set("Erreur : veuillez entrer un rayon")
            return

        rayon = float(rayon_str)
        c = Cercle(rayon)
        surface = c.calculer_surface()
        resultat_var.set(f"Surface : {surface:.2f}")

    except ValueError:
        resultat_var.set("Erreur : rayon invalide")


def reset():
    rayon_entry.delete(0, END)
    resultat_var.set("Surface : en attente...")



calculer_btn = Button(Fen, text="Calculer la surface", bg="#87CEEB", fg="white", command=calculer)
calculer_btn.place(x=80, y=90)

reset_btn = Button(Fen, text="Effacer", bg="#87CEEB", fg="white", command=reset)
reset_btn.place(x=250, y=90)

exit_btn = Button(Fen, text="Quitter", command=Fen.destroy, bg="#87CEEB", fg="white")
exit_btn.place(x=320, y=150)
```
