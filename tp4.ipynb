```python
import os
from random import sample #sample permet de tirer un échantillion aléatoire à partir d'une séquence sans répétition 
with open ('contact.txt','r') as f:
    lignes=f.readlines() #ligne de contenu
    for ligne in lignes: #lecture element par element
        print(ligne)
        L=sample(ligne,len(ligne))
        print(L)
        ch=''
        for elt in L:
            if elt ==' ' or elt=='\n' or elt==',':
                pass
            else:
                ch+=elt
        print(ch)
        print('______________')
        if os.path.exists('motdepass.txt'):
            print('file exists')
        else:

            with open('motdepass.txt','a') as motpass:
                motpass.write(ch+'\n')
```
